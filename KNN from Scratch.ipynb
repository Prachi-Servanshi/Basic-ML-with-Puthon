```python
import math

data = pd.read_csv("iris.csv", header=None)

#print(data)

k = int(input("Enter a Value of K: "))
test = [4.5, 2.3, 1.3, 0.3]

dist_data = list()

'''for row in data.values:
    for i in range(len(row)):
        print(row[i])'''

for row in data.values:
    distance = 0.0
    for i in range(len(row) - 1):
        x1 = float(test[i])
        x2 = float(row[i])
        distance = distance + ((x1 - x2) * (x1 - x2)) 
    dist = math.sqrt(distance)
    dist_data.append((row, dist))


def funckey(arr):
    return arr[1]


dist_data.sort(key = funckey)

#print(dist_data)

knearest = list()
for i in range(k):
    knearest.append(dist_data[i][0])


#print(knearest)


nearestclass = list()

for x in knearest:
    nearestclass.append(x[4])

        
check = nearestclass[0]
#print(check)
maxcnt = nearestclass.count(check)
prediction = nearestclass[0]
#print(maxcnt)
for x in nearestclass:
    count = x.count(x[4])
    if (count > maxcnt):
        maxcnt = count
        prediction = x[4]


print("The Predicted Class for ",test,"is: ",prediction)
```
